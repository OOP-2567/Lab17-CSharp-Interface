# Lab 17 Exercise 1

## Interface

1.สร้าง console application project

```cmd
dotnet new console --name Lab17_Ex01
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var animal_1 = new Dog();
animal_1.Move();

IMovable animal_2 = new Dog();
animal_2.Move();

interface IMovable
{
    public void Move();
}

abstract class Animal : IMovable
{
    abstract public void Move();
}
class Dog : Animal
{
    public override void Move()
    {
        System.Console.WriteLine("Dog move by running on the ground");
    }
}
```

3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab17_Ex01
```

ถ้ามีที่ผิดพลาดในโปรแกรม ให้แก้ไขให้ถูกต้อง

4.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 3

การใช้ Polymorphism ช่วยให้เรียกเมธอด Move() ได้แม้จะอ้างถึงออบเจกต์ผ่านอินเทอร์เฟซ การสืบทอดจากคลาส Animal บังคับให้ Dog ต้อง override เมธอด Move() เพื่อให้ทำงานได้ตามที่ต้องการ
