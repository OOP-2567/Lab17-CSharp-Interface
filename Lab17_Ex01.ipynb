# Lab 17 Exercise 1

## Interface

1.สร้าง console application project

```cmd
dotnet new console --name Lab17_Ex01
```

2.เปลี่ยน code ให้เป็นดังต่อไปนี้

```csharp
var animal_1 = new Dog();
animal_1.Move();

IMovable animal_2 = new Dog();
animal_2.Move();

interface IMovable
{
    public void Move();
}

abstract class Animal : IMovable
{
    abstract public void Move();
}
class Dog : Animal
{
    public override void Move()
    {
        System.Console.WriteLine("Dog move by running on the ground");
    }
}
```

3.Build project โดยการใช้คำสั่ง

```cmd
dotnet build  Lab17_Ex01
```

6.บันทึกผลที่ได้จากการรันคำสั่งในข้อ 5

7.อธิบายสิ่งที่พบในการทดลอง

- Interface (IMovable): กำหนดให้คลาสที่ implement interface นี้ต้องมี method Move(). โดย method นี้จะไม่มีการกำหนดการทำงานใน interface แต่จะเป็นการบังคับให้คลาสที่ใช้งาน interface นี้ต้องมีการ implement method ดังกล่าว
- Abstract Class (Animal): ใช้เป็นคลาสแม่ที่ implement interface IMovable และกำหนด method Move() เป็น abstract method คลาสลูกที่สืบทอดจาก Animal จะต้อง implement method นี้
- คลาส Dog: สืบทอดจาก Animal และ override method Move() เพื่อให้มีพฤติกรรมการเคลื่อนที่ของสุนัข
- Polymorphism: สามารถใช้ interface เพื่อเก็บอ็อบเจ็กต์ของคลาสที่ implement interface นั้น ๆ และเรียกใช้ method ที่ถูก implement โดยไม่ต้องรู้ว่าคลาสนั้น ๆ เป็นคลาสใด
